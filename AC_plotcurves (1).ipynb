```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
import pandas as pd
import csv
import sys
import os
from scipy.optimize import curve_fit
from scipy import optimize
```

```python
header_row=['x','y']
ch1_pointy = pd.read_csv("channel1_AC_1 - Copy.csv", names = header_row)
ch2_pointy = pd.read_csv("channel2_AC_1 - Copy.csv", names = header_row)
ch1_S = pd.read_csv("channel1_AC_2 - Copy.csv", names = header_row)
ch2_S = pd.read_csv("channel2_AC_2 - Copy.csv", names = header_row)
ch1_pointy_NR = pd.read_csv("ch1_pointy_normalR - Copy.csv", names = header_row)
ch2_pointy_NR = pd.read_csv("ch2_pointy_normalR - Copy.csv", names = header_row)
ch1_VV = pd.read_csv("ch1_VV_normalR - Copy.csv", names = header_row)
ch2_VV = pd.read_csv("ch2_VV_normalR - Copy.csv", names = header_row)
```

```python
orange = '#f46d25'

df1 = ch1_pointy

NMR = df1.drop(0)

Ch1t = NMR.loc[:, "x"]

Ch1V = NMR.loc[:, "y"]
#plt.figure(figsize=(20,20))
#plt.plot(Ch1t, Ch1V, marker = 's')

df2 = ch2_pointy

NMR = df2.drop(0)

Ch2t = NMR.loc[:, "x"]

Ch2V = NMR.loc[:, "y"]
#plt.figure(figsize=(20,20))
#plt.plot(Ch2t, Ch2V, marker = 's')

def test_func(x, a, b):
    return a * x + b

params, params_covariance = optimize.curve_fit(test_func, Ch2V, Ch1V,p0=[2, 2])
a=params[0]
b=params[1]

plt.figure(figsize=(10,10))
plt.plot(Ch2V,Ch1V,c=orange)
plt.title("Resistor Voltage Input/Output Comparison")
plt.xlabel('Input Voltage (V)')
plt.ylabel('Output Voltage (V)')
plt.plot(Ch2V, test_func(Ch2V, params[0], params[1]),label='Fitted function (Output Voltage = Input Voltage * %.3f + %.3f)' % (a, b))
plt.legend(loc='best')
```
